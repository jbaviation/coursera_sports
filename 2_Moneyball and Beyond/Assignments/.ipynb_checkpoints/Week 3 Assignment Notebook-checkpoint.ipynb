```python
# First we load the packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.formula.api as smf
```

# I. Data Preparation

```python
# Step (2)
#  Read in master data
#  Note this master dataset was created in the Week 2 assignment
Master = pd.read_csv("../Data/Master.csv")
master = Master.iloc[:,1:]
```

```python
# Step (3)
#  Create “Experience Squared” variable  
master['exp2'] = master['Exp']**2
master.iloc[0,:]
```

```python
# Step (4)
#  Calculate variables for batting average, isolated power, and eye as defined above (for “eye” do not 
#  include IBB in walks)
master['BA'] = master['H'] / master['AB']
master['IsoPow'] = master['SLG'] - master['BA']
master['eye'] = (master['BB']+master['HBP'])/master['PA']
master
```

```python
# Step (5)
#  Subset the data to only include seasons (SalYear) 1995-2015
master = master[(master['SalYear']>=1995) & (master['SalYear']<=2015)]
master
```

```python
# Quiz
# Q1
master.sort_values('eye', ascending=False)[['playerID','yearID','eye']]
```

```python
# Q2
# master['IsoPow']
master.groupby(['teamID','yearID'])['IsoPow'].mean().reset_index().sort_values('IsoPow',ascending=False)
```

```python
# Q3
master.groupby('yearID')['BA'].median().sort_values(ascending=False)
```

# II. Running Regressions for Each Season

```python
# Step (1)
#  Write a function to run the Moneyball regression annually for free agents only
def MBExpandFA(Season, df=master):
    mb_sea = df[(df.SalYear == Season) & (df.Free == 1)]
    return smf.ols(formula='lnSal~BA+IsoPow+eye+PA+Exp+exp2+C(POS)', data=mb_sea).fit()
```

```python
# Steps (2)-(4)
seasons = list(range(1995, 2015))
lm_res = [MBExpandFA(sea) for sea in seasons]
```

```python
# Step (5)
pre_mb = lm_res[:seasons.index(2001)]
mb_era = lm_res[seasons.index(2001):seasons.index(2009)]
post_mb= lm_res[seasons.index(2009):]
```

```python
# Step (6)
from statsmodels.iolib.summary2 import summary_col
for era, years in zip([pre_mb,mb_era,post_mb], [range(1995,2001),range(2001,2009),range(2009,2015)]):
    print(summary_col(era, model_names=[str(s) for s in years], stars=True, 
                      info_dict={'Num Observations': lambda x: f'{int(x.nobs):d}'}))
```
