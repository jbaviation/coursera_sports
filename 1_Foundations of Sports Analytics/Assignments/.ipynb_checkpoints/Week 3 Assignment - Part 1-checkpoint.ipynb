## Part 1 - NBA Player Comparision

```python
# As usual, we begin by importing the packages we will need

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# The data consists of the shot log for the NBA season 2016/17

shot = pd.read_csv("Assignment Data/NBA shotlog_16_17.csv")
pd.set_option('display.max_columns', 100)
print(shot.columns.tolist())
shot.describe()
```

## Assignment Instructions
1. Load the data containing the NBA shot log for the 2016/17 season that we used earlier
2. Use the same code that we used to project all shots onto a half court by defining shot[‘halfcourt_x’] and shot[‘halfcourt_y’]
3. Now define subsets for the following players: Kevin Durant, Dwight Howard, DeAndre Jordan and Russell Westbrook.
4. Create plots of their shots in the same way that we did for Steph Curry and LeBron James: copy the code we used and just change the names. Show the plots of Russell Westbrook and Kevin Durant side by side. In order to make sure that the two plots have the same ranges, for each subplot add the lines:
   - plt.xlim(500,950)
   - plt.ylim(0,500)
5. Create the plot of DeAndre Jordan and Dwight Howard side by side
6. Create the plot of Brook Lopez and Robin Lopez side by side

```python
# (2)
shot['halfcourt_x'] =np.where(shot['location_x'] < 933/2, 933 - shot['location_x'],shot['location_x'])
shot['halfcourt_y'] =np.where(shot['location_x'] < 933/2, 500 - shot['location_y'],shot['location_y'])
```

```python
# (3)
players = ['Kevin Durant', 'Dwight Howard', 'DeAndre Jordan', 'Russell Westbrook']
durant = shot[shot['shoot_player']=='Kevin Durant']
howard = shot[shot['shoot_player']=='Dwight Howard']
jordan = shot[shot['shoot_player']=='DeAndre Jordan']
westbrook = shot[shot['shoot_player']=='Russell Westbrook']
brooklopez = shot[shot['shoot_player']=='Brook Lopez']
robinlopez = shot[shot['shoot_player']=='Robin Lopez']
```

```python
# (4)
def side_side(player1, player2, name1, name2):
    f = plt.figure(figsize=(94/6,50/6))
    
    # Player 1
    ax = f.add_subplot(121)
    colors = np.where(player1['current_shot_outcome']=='SCORED','r',np.where(player1['current_shot_outcome']=='MISSED','b','g'))
    ax = plt.scatter(player1['halfcourt_x'],player1['halfcourt_y'], s=10, c= colors, marker= '.')
    plt.grid(True)
    plt.title(name1, fontsize = 15)
    plt.xlim(500,950)
    plt.ylim(0,500)
    
    # Player 2
    ax = f.add_subplot(122)
    colors = np.where(player2['current_shot_outcome']=='SCORED','r',np.where(player2['current_shot_outcome']=='MISSED','b','g'))
    ax = plt.scatter(player2['halfcourt_x'],player2['halfcourt_y'], s=10, c= colors, marker= '.')
    plt.grid(True)
    plt.title(name2, fontsize = 15)
    plt.xlim(500,950)
    plt.ylim(0,500)
```

## Westbrook - Durant Comparision

```python
side_side(westbrook, durant, players[3], players[0])
```

## Jordan - Howard Comparision

```python
side_side(jordan, howard, players[2], players[1])
```

## Lopez - Lopez Comparision

```python
side_side(brooklopez, robinlopez, 'Brook Lopez', 'Robin Lopez')
```
