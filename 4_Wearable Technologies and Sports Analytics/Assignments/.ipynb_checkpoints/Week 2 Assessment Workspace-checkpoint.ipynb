## Wearable Tech: Week 2 Assignment

### Step 1

```python
# imports
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import datetime
import seaborn as sns

#turns off warning for chained assignment as we are overwriting references to original dataframe
pd.options.mode.chained_assignment = None  # default='warn'
pd.set_option('display.max_rows', 30)

df = pd.read_csv("../Data/Soccer.csv", parse_dates=["Date"])
```

### Step 2

```python
soccer = df[df['Period.Name']=='Session'].copy()
soccer
```

### Step 3

```python
soccer['duration'] = soccer['Total.Player.Load'] / soccer['Player.Load.Per.Minute']
soccer
```

### Step 4

```python
soccer['Player.Name'] = soccer['Player.Name'].astype('object')
soccer.groupby('Player.Name')['Total.Player.Load'].sum().astype('int').sort_values(ascending=False)
```

```python
soccer[soccer['Player.Name']==98]
```

### Step 5

```python
soccer.set_index('Player.Name',drop=True).sort_values('Total.Player.Load', ascending=False).head(10)
# all occurred on game days
```

### Step 6

```python
drop_ = ['98','62','89','23','71']
soccer_sub = soccer[~soccer.isin(drop_)].copy()
soccer_sub.head()
```

### Step 7 

```python
# Player 43 is max in Step (4)
df_plot = soccer[soccer['Player.Name']==43]

# Set datetime
df_plot['Date'] = pd.to_datetime(df_plot['Date'])
df_plot = df_plot.set_index('Date', drop=True)

# Calculate acute:chronic
df_plot['acute_pl'] = df_plot['Total.Player.Load'].rolling('7D').mean()
df_plot['chronic_pl'] = df_plot['Total.Player.Load'].rolling('28D').mean()
df_plot['acwr'] = df_plot['acute_pl'] / df_plot['chronic_pl']

# Plot
fig, ax = plt.subplots(figsize=(20,10))
ax.set_xlabel('Date')
ax.set_ylabel('ACWR')
ax.plot_date(df_plot.index, df_plot['acwr'], xdate=True)
# fig.autofmt_xdate()
# ax.set_xlim([datetime.date(2019,8,6), datetime.date(2019,11,25)])
```

### Step 8

```python
# Plot Whole Team
soccer['Date'] = pd.to_datetime(soccer['Date'])
soccer_team = soccer.groupby('Date').mean()

# Calculate acute:chronic
soccer_team['acute_pl'] = soccer_team['Total.Player.Load'].rolling('7D').mean()
soccer_team['chronic_pl'] = soccer_team['Total.Player.Load'].rolling('28D').mean()
soccer_team['acwr'] = soccer_team['acute_pl'] / soccer_team['chronic_pl']

# Plot
fig, ax = plt.subplots(figsize=(20,10))
ax.set_xlabel('Date')
ax.set_ylabel('ACWR')
ax.plot_date(soccer_team.index, soccer_team['acwr'], xdate=True)
```
