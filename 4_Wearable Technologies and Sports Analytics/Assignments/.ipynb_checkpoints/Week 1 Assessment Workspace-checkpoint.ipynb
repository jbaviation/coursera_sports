## Wearable Tech: Week 1 Assignment

## Step 1

```python
#imports
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
import datetime
%matplotlib inline

# Read in the dataset

df = pd.read_csv("../Data/TexasTechVBall.csv")
```

## Step 2

```python
df_vb = df.copy()
df_vb['DATE'] = pd.to_datetime(df_vb['DATE'], errors='coerce')
df_vb['WEEK'] = (df_vb['DATE'].dt.isocalendar().week - 33).astype('object')
df_vb.head()
```

## Step 3

```python
df_vb.groupby(['EVENT TYPE','POSITION'], as_index=False)[['JUMPS','JUMPS 15+','JUMPS 20+','HIGHEST JUMP (in)']].mean().sort_values('HIGHEST JUMP (in)')
```

## Step 4

```python
df_vb.groupby(['EVENT TYPE','POSITION'], as_index=False)[['ELEVATED LANDING %','HIGH LANDING %','ALERT LANDING %']].mean()
```

## Step 5

```python
sns.set(rc={'figure.figsize': (15,8)})
sns.lineplot(y='ELEVATED LANDING %', x='WEEK', data=df_vb, hue='PLAYER NAME', ci=None)
```

## Step 6

```python
el_pct = df_vb.corr()['ELEVATED LANDING %'].sort_values().to_frame()
plt.figure(figsize=(4,6))
sns.heatmap(el_pct, cbar=False, annot=True, linewidths=0.5, cmap="rocket")
```
