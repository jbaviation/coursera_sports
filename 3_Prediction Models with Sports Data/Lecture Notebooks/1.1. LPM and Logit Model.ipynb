### Linear Probability Model (LPM)
LPM is the regression model applied to a binary dependent variable:

$$
\begin{align}
Y &= a+bX+\epsilon \\
Y &= 1\text{, when some events occur} \\
Y &= 0\text{, when some events don't occur}
\end{align}
$$

where `Y` is a **binary random variable** which varies from 0->1.  Recall from a previous course that this is of the same form as the linear regression model with **one key difference**; the nature of the dependent variable.  More specifically, the dependent variable is **continuous** in the **linear regression model**, whereas the dependent variable is **binary** in the **LPM**.  The LPM can also be thought of as a special linear regression case with a binary result.  The conditional expectations are represented by:

$$
Pr(Y=1|X) = Xb
$$

#### Interpretation of LPM
Given the following equation taken from an analysis that will be performed later:

$$ E(Y=1|X) = -0.11 + 1.22*(Pyth\%) $$

This equation can be interpretted as: For every ***1% increase*** of Pythagorean winning %, the expected chance of winning increases by ***1.22%***.

### 1.4) Add Pythagorean winning % to "NHL_reg_2017" dataset, named "pyth_wpct" 

Now we need to add Pythagorean Winning percentage to be used as I.V for regression. In order to obtain Pythagorean winning percentages, we need to obtain cumulative GF and GA for each team in each game. 

```python
## Add Pythagorean Win-percent
#1) Sort the dataframe by game in order (i.e., gid)
NHL_reg_2017 = NHL_reg_2017.sort_values(by ='gid')

#2) Calculate cumulative GF and GA for each team
NHL_reg_2017['cumGF'] = NHL_reg_2017.groupby(['tid'])['goals_for'].apply(lambda x: x.cumsum())
NHL_reg_2017['cumGA'] = NHL_reg_2017.groupby(['tid'])['goals_against'].apply(lambda x: x.cumsum())

display(NHL_reg_2017)
```

```python
#Create Pythagorean win-percent
NHL_reg_2017['pyth_wpct'] = NHL_reg_2017['cumGF']**2/(NHL_reg_2017['cumGF']**2 + NHL_reg_2017['cumGA']**2)

display(NHL_reg_2017[0:10])
```

## 3. Logistic Regression
The functional form of the logistic regression is known as the **Log of the Odds**, aka **logit**.  The logit function is used to resolve issues that occur with LPM as represented in the following formula: 

$$ \ln\left[\frac{Pr(y=1|x)}{1-Pr(y=1|x)}\right] = \alpha + \beta x $$

#### Log of the Odds
This term refers to *odds* as-in a comparison to the probability. Note the differences between *odds* and *probability* below.

The **probability** indicates how often certain event(s) happen relative to the total events that would occur.

$$P(A) = \frac{\text{Certain Events}}{\text{Total Events}} $$

The **odds** indicate how often something happens relative to how often it does not happen:

$$\frac{P(A)}{1-P(A)}$$

#### Comparison of Probability vs Odds vs Logit
The following table shows what the corresponding values are:

| Probability | Odds | Logit |
| --- | :---: | ---: |
| 0.01 | 1/99=0.0101 | -4.60 |
| 0.05 | 5/95=0.0526 | -2.94 |
| 0.30 | 3/7=0.4286 | -0.85 |
| 0.70 | 7/3=2.333 | 0.85 |
| 0.92 | 95/9=19 | 2.94 |
| 0.99 | 99/1=99 | 4.60 |

#### Ideas of Logistic Regression
The probability is transformed into **odds*:
$$ \text{Odds} = \frac{Pr(y=1|x)}{Pr(y=0|x)} = \frac{Pr(y=1|x)}{1-Pr(y=1|x)} $$

Then we take the **log** of the odds:
$$\ln\left[\frac{Pr(y=1|x)}{1-Pr(y=1|x)}\right]$$

The we use the **logit** as a dependent variable:
$$\ln\left[\frac{Pr(y=1|x)}{1-Pr(y=1|x)}\right] = \alpha + \beta x$$

Using the more exact definition of probability, we know that the probability must always be positive:
$$Pr(y=1|x) = \exp(\alpha+\beta x) = e^{\alpha+\beta x}$$

And must be less than 1:
$$Pr(y=1|x) = \frac{e^{\alpha+\beta x}}{e^{\alpha+\beta x}+1} = \frac{1}{e^{-(\alpha+\beta x)}+1}$$

#### Interpretation of Coefficients
The logit model is a linear function of x:
$$\ln\left[\frac{Pr(y=1|x)}{1-Pr(y=1|x)}\right] = -2.75 + 5.48*\text{Pyth%}$$

The *dependent variable* takes a form of logit, which can make it difficult to interpret the impact of the *independent variable*.  This can be written now as:
$$\ln\left[\frac{Pr(y=1|x)}{1-Pr(y=1|x)}\right] = \frac{1}{\exp(-(-2.75+5.48*\text{Pyth%}))+1}$$

- a Pyth% of 0.2, the chance of winning is 0.16
- a Pyth% of 0.3, the chance of winning is 0.24
- a Pyth% of 0.4, the chance of winning is 0.36

### 3.1) Import libraries 

```python
from sklearn.linear_model import LogisticRegression
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

### 3.2) Fit a logistic regression of win on pyth_wpct & print outcomes

```python
Win_Pyth = 'win~pyth_wpct'
model = smf.glm(formula = Win_Pyth, data = NHL_reg_2017_WL, family=sm.families.Binomial())
result = model.fit()
print(result.summary())
```

```python
print("Coefficients")
print(result.params)
print("p-Values")
print(result.pvalues)
print("Dependent variables")
print(result.model.endog_names)
```

### 3.3) Obtain the fitted probabilities of winning on each game by using the model

```python
# we can fit the model to calculate probabilities of winning on each game
## Here we can print first 10 probabilities corresponding to the probabilities of winning
fittedProbs = result.predict()
print(fittedProbs[0:10])
```

### 3.4) Create a binary winning variable by using the fitted probabilities

```python
# Based on the fitted probabilites of winning, here we create a binary winning variable for 1 indicating Win (Prob. > 50%), 0 indicating lose (Prob. < 50%)
fittedWin = [1 if x > .5 else 0 for x in fittedProbs]
print(fittedWin[0:10])
```

### 3.5) Measuring model performance: Confusion matrix

#### A confusion matrix is a table that is often used to describe the performance of a logistic regression model on a set of data for which the true values are known. 

```python
from sklearn.metrics import confusion_matrix, classification_report
confusion_matrix(NHL_reg_2017_WL['win'], fittedWin)
```

```python
# Success rate
(736+745)/2454
```

```python
print(classification_report(NHL_reg_2017_WL['win'], fittedWin, digits=3))
```

## 5. Discussion: Statistical Forecasting and It's Real World Implication

The everyday meaning of predictions is to state the expected outcome of an event before it happens, it means the same as "forecast". whereas, in statistics, it is often used to mean simply the fitted value from a model. In our example, the model using Pyth.win% cannot give us a prediction in the everyday sense. (think about this per second)
It is because that we need to know the Goals For and Against, which happen at the same time as the game. In other words, In a real timeframe, it's not possible to use pyth.Win% to predict the game results. 

For our model to be predictive for future events, we need a plausible independent variable, which can be obtained beforehand.
